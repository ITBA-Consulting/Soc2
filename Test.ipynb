```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
@interact
def show_articles_more_than(column='Battery V', x=4):
    display(HTML(f'<h2>Showing articles with more than {x} {column}<h2>'))
    display(df.loc[df[column] > x, ['Date and Time','Date (MM/DD/YYYY)','Time (HH:MM:SS)','Date (MM/DD/YYYY)','Time (HH:MM:SS)','Battery V','Temp °C','Cond µS/cm','SpCond µS/cm','Sal psu','nLF Cond µS/cm','ODO % sat','ODO mg/L','Turbidity FNU','TSS mg/L','pH','pH mV','ORP mV','ORP raw mV','Press psi a','Depth m']])
```

```python
@interact
def correlations(column1=list(df.select_dtypes('number').columns), 
                 column2=list(df.select_dtypes('number').columns)):
    print(f"Correlation: {df[column1].corr(df[column2])}")
```

```python

@interact
def line_plot(x=list(df.select_dtypes('number').columns), 
                 y=list(df.select_dtypes('number').columns)[1:]):
    df.iplot(kind='line', x=x, y=y, mode='lines', 
             xTitle=x.title(), yTitle=y.title(), title=f'{y.title()} vs {x.title()}')
```

```python
import cufflinks as cf

@interact
def scatter_plot(x=list(df.select_dtypes('number').columns), 
                 y=list(df.select_dtypes('number').columns)[1:],
                 theme=list(cf.themes.THEMES.keys()), 
                 colorscale=list(cf.colors._scales_names.keys())):
    
    df.iplot(kind='scatter', x=x, y=y, mode='markers', 
             xTitle=x.title(), yTitle=y.title(), 
             title=f'{y.title()} vs {x.title()}',
            theme=theme, colorscale=colorscale,
            dimensions=(1000, 750))
```
